```python
import pandas as pd
file_path='/content/drive/MyDrive/bank_train.csv'
df=pd.read_csv(file_path)
print(df)
```

```python
df.head(),df.columns
```

```python
primary_edu_deposit = df.loc[(df['education'] == 'primary') & (df['deposit'] == 'yes')]
primary_edu_deposit.head()
```

```python
not_subscribed=df.loc[df['deposit']=='no']
not_subscribed.head()
```

```python
subscribed_with_loans = df.loc[(df['deposit'] == 'yes') & ((df['housing'] == 'yes') | (df['loan'] == 'yes'))]
subscribed_with_loans.head()
```

```python
secondary_edu_not_subscribed = df.loc[(df['education'] == 'secondary') & (df['deposit'] == 'no')]
secondary_edu_not_subscribed.head()
```

```python
subscribed_successful_campaign = df.loc[(df['deposit'] == 'yes') & (df['poutcome'] == 'success')]
subscribed_successful_campaign.head()
```

```python
unemployed_not_subscribed = df.loc[(df['job'] == 'unemployed') & (df['deposit'] == 'no')]
unemployed_not_subscribed.head()
```

```python
young_clients_salary = df.loc[df['age'] <= 30, ['age', 'balance']]
young_clients_salary.head()
```
